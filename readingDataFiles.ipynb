```python
import os
import numpy as np

import pydicom
import SimpleITK as sitk
import matplotlib.pyplot as plt
```

```python
dicom_file_path = 'zurich_01/CT2.16.840.1.114493.1.4.228.5.20220617143709187.dcm'
dicom_data = pydicom.dcmread(dicom_file_path)

# data info
pixel_array = dicom_data.pixel_array
print(pixel_array.shape)
# print(dicom_data)

# data display
plt.imshow(pixel_array, cmap='gray')
plt.show()
```

```python
dicom_dir = 'zurich_01'
dicom_files = [os.path.join(dicom_dir, f) for f in os.listdir(dicom_dir) if f.endswith('.dcm')]

dicom_files.sort()   ### ?? do I need this line? what sorting order matters for my work?
dicom_files = dicom_files[150:155]

zurich_01 = []
# Read and display each DICOM file
for dicom_file in dicom_files:
    dicom_data = pydicom.dcmread(dicom_file)
    pixel_array = dicom_data.pixel_array

    # plt.imshow(pixel_array, cmap='gray')
    # plt.title(dicom_file)
    # plt.show()
    zurich_01.append(pixel_array)

zurich_01 = np.array(zurich_01)
print(zurich_01.shape)
```

```python
# Replace 'path/to/your/file.mha' with the actual file path
mha_file_path = 'PRT_001_Fx5/masks/PTV_3CM.mha'  # PRV: is the prostate volume, PTV_2CM and PTV_3CM: are the planning target volumes

# Read the .mha file
image = sitk.ReadImage(mha_file_path)

# Get the numpy array from the SimpleITK image
image_array = sitk.GetArrayFromImage(image)
print(image_array.shape)
print(image_array[143].shape)

# Print basic information about the image
print(f"Image size: {image.GetSize()}")
print(f"Image spacing: {image.GetSpacing()}")
print(f"Image direction: {image.GetDirection()}")
print(f"Image origin: {image.GetOrigin()}")


for i in range(60,65):
    plt.imshow(image_array[i], cmap='gray')
    plt.title("mha_file")
    plt.show()
```
